```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load the diabetes dataset
diabetes = datasets.load_diabetes()
X = diabetes.data[:, np.newaxis, 2]  # Use one feature (e.g., BMI)
Y = diabetes.target  # Target variable

# Split the data into training/testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)

# Predictions
y_test_pred = model.predict(X_test)

# Plot the results
plt.figure(figsize=(5, 3))
plt.scatter(X_test, Y_test, color="blue", label="Actual data")
plt.plot(X_test, y_test_pred, color="red", label="Regression line")
plt.title("Linear Regression on Diabetes Dataset(Testing Data)")
plt.xlabel("BMI")
plt.ylabel("Diabetes Progression")
plt.legend()
plt.show()
```
